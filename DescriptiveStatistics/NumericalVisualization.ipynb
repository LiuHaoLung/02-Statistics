Each number occurs exactly one time, this tabel would be impractical for any analysis right.

When we deal with numerical variabels it makes much more sence to group the data into intervals and find the corresponding frequencies.

We will divide the data into five intervals and the interval width formula we use is as follow : (largest number - smallest number) / number of desired intervals = (100 - 1) / 5 = 19.8 and we want to round this number up in order to reach a need or representation, so the interval width is 20.

So the interval table is below:

---                |---

21                 |41
41                 |61
61                 |81
81                 |101

So construct the frequency distribution tabel is:

---                |---             |---
1                  |21              |2

41                 |61              |3
61                 |81              |6
81                 |101             |5

The rule of a number is included in an interval if that number:

1. is greater than the lower bound.
2. is lower or equal to the upper bound.

The most common graph used to represent numerical data is the histogram.

The histogram is looks like a bar chart but actually conveys very different information.

The vertical axis is a numerical type and shows the absolute frequency, the horizontal is numberical too.

So each bar has with equal to the interval and height equal to the frequency.

The different bars are touching, this is show there is continuation between the intervals, each interval ends where the next one starts.
