In the glassdoor the salary is 113000, the sample that is available on glassdoor based on self-reported and you would like to see if its value is correct.

We need a two sided test as we are interested in knowing both that the salary is significantly less or significantly more than that.

The null hypothesis(<code>H0:μ0</code>) is the population mean salary is 113000, the alternative hypothesis(<code>H1:μ0</code>) is the population mean salary is different than 113000.

Testing is done by standardizing the variable at hand and comparing it to the Z score.

We standardize a variable by subtracting the mean and dividing by the std, since it is a sample we use the standard error.

Thus the formula for standardization becomes <code>Z = (x̄ - μ0) / (σ / (n ^ 0.5))</code>, in this way we obtain a distributiom with a mean of 0 and std is 1.

Z means standardized varaible associated with the test called the Z score; z means one from the table and will be referred to as the critical value.

The z is normally distribution with a mean and std of 1, that means <code>Z ~ N(0,1)</code>; the Z is normally distributed with mean of <code>x̄ - μ0</code> and std, that means <code>Z ~ N(x̄ - μ0,1).</code>

Standardization let us compare the means, the closer different of <code>x̄</code> and <code>μ0</code> to 0, the closer the z score itself to 0, this imply the higher chance to accept the null hypothesis.

Standardizied variable, we put the value into the formula is <code>(100200 - 113000) / 2739 = -4.67</code>, now we will compare absolute value of -4.67 with a <code>α/2</code> where the <code>α</code> is the significance level.

Some z tables don't include negative values, as the standard normal distribution is symmetrical around 0, the two statements are equivalent <code> -4,67 < a negative z = 4.67 > a positive z.</code>
    
So the decision rule is become: if we want to reject it, the rule is absolute value of Z-score > positive critical value(z).

Using the 5% significance, our <code>α = 0.05</code>, since it is a two sided test, we check the table of <code>Z0.025 = 1.96.</code>

Our Z us 4.67 and it is higher than 1.96, so we reject the null hypothesis, the answer at 5% significance level we have reject the null hypothesis or at 5% significance level there is no statistical evidence that the mean salary is 113000.
