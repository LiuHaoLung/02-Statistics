# Test For Mean & Independent Samples

#### It is independent samples with known variance

Example we would like to test the grades of two departments in a UK university, the dean say that engineering is a tougher discipline than people tend to get lower grades, he believes that on averager management students outperform engineering by 4%.

The null hypothesis is <code>H0:μe - μm = -4%</code>; the alternative hypothesis is <code>H1:μe - μm ≠ -4%.</code>

Our research question is not defined the difference, but to check if it is exactly for, so it is two-sided.

The summarizes the data table is below:

Type            |Engineering(x)       |Management(y)     |Difference(x - y)
---             |---                  |---               |---
Size            |100                  |70                |?
Sample mean     |58                   |65                |-7.0
Population std  |10                   |6                 |1.23(std error)

We have big samples and known variances, so we use the Z-score, but if it have big samples and unknoun variances it i up to the researcher.

The Z-score is <code>((-7%) - (-4%)) / 1.23 = -2.44.</code>

The other value is below:

Type            |Values
---             |---
Z-score         |-2.44
p-value         |0.015

So if the level of significance is 5% and the p-value is 0.015, we can reject the null hypothesis, that means there is enough statistical evidence that the mean difference is not 4%.

How about is the differenct higher or lower than 4% ? The sign of the test statistic shows if the mean is lower or higher than the hypothesized value.

The negative sign of statistic means it is smaller, if reverse engineer the standardization process, you will find that true value is likely to be lower than the hypothesized value, in our case this translates into the true mean is likely to be lower than -4%.

#### It is independent samples with unknown variance but assumed to be equal

Example you were trying to see if apples in NY are as expensive as the ones in LA, you want to 10 grocery shops in NY and your friend lives in LA went to 8 grocery shops there, you got all the prices and put them in a table, what the population variance of apples prices is but you assume it should be the same for NY and LA.

The null hypothesis is <code>H0:μNY = μLA</code>, so the null hypothesis is <code>H0:μNY - μLA = 0.</code>

The alternative hypothesis is <code>H1:μNY ≠ μLA</code>, so the alternative hypothesis is <code>H1:μNY - μLA ≠ 0.</code>

The new summarizes the data table is below:

Type            |NY(x)      |LA(y)

Mean            |3.94       |3.25
Std. deviation  |0.18       |0.27
Sample size     |10         |8

The pooled variance formula is [((x's sample size - 1) * (std. deviation)^2) + ((y's sample size - 1) * (std. deviation)^2)] / (x's sample size + y's sample size - 2) = <code>[(10 - 1)*(0.18 ^2) + (8 - 1) *(0.27^2)] / (10 + 8 - 2) = 0.05.</code>

The pooled std formula is [(pooled variance) / (sample size) + (pooled variance) / (sample size)] ^ 0.5 = <code>[(0.05 / 10) + (0.05 / 8)] ^ 0.5 = 0.11.</code>


So the new table is below:

pooled std         |0.11       |0.11

We have small samples and unknown variances, so in this case we need use the t-score and the degrees for freedom is <code>10 + 8 - 2 = 16.</code> and the t-score is 6.53.

For common tests a rule of thumb is to reject the null hypothesis when the T-score is bigger than 2.

Generally, for Z and T-score, a value higher than 4 is extremely significant.
